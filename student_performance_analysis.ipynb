```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in Cat_col:
    plt.figure(figsize=(7, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df.corr(numeric_only=True),kde=True,bins=30,palette='Set2')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Flatten the correlation matrix and remove self-correlations (i.e., 1s on the diagonal)
corr_matrix = df.corr(numeric_only=True)
corr_values = corr_matrix.where(~np.eye(corr_matrix.shape[0],dtype=bool)).stack()

plt.figure(figsize=(10, 6))
sns.histplot(corr_values, kde=True, bins=30, color='skyblue')
plt.title('Distribution of Correlation Values (Excluding Diagonal)')
plt.xlabel('Correlation')
plt.ylabel('Frequency')
plt.show()
```

```python


for col in Cat_col_int:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color='salmon')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

    # Boxplot to check outliers
    plt.figure(figsize=(6, 1))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
df['Average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
def performance_lable(score):
    if score >= 90:
        return 'Very Good'
    elif score >= 70:
        return 'Good'
    else:
        return 'Poor'
df['Score_lable']=df['Average_score'].apply(performance_lable)
plt.figure(figsize=(10,8))
sns.countplot(x='Score_lable',data=df,hue='gender')
plt.show()
```

```python

# Ensure consistent styling
sns.set_theme(style="whitegrid")

# Step 1: Feature engineering â€“ if not already done
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Step 2: Group by multiple categories
grouped_df = (
    df.groupby(['gender', 'test preparation course', 'parental level of education'])
      .agg(average_score=('average_score', 'mean'),
           count=('average_score', 'count'))
      .reset_index()
      .sort_values(by='average_score', ascending=False)
)

# Step 3: Filter to show only groups with at least 20 students
filtered_df = grouped_df[grouped_df['count'] >= 20]

# Step 4: Pivot table for easier visualization
pivot = filtered_df.pivot_table(index='parental level of education',
                                 columns=['gender', 'test preparation course'],
                                 values='average_score')

# Step 5: Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title('Average Student Score by Parental Education, Gender, and Prep Course')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
